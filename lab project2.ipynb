# Matplotlib example workflows




```python
# 0,import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt
#1,prepare data
x=[1,2,3,4]
y=[11,22,33,44]
#2,setup plot
fig,ax=plt.subplots(figsize=(5,6))
#3,plot data
ax.plot(x,y)
#4,customize data
ax.set(title ="simple plot",
       xlabel="x-axis",
       ylabel ="y-axis")
#5,save and show the whole figure
plt.savefig('figures/sample -plot.png')
```

# # importing csv file and performing visualization


```python
df=pd.read_csv("googleplaystore.csv")
df
```

```python
df["Rating"]=df["Rating"].replace(np.NaN,0)
df
```

```python
del df['Price']
df
```

```python
df["Size"]=df["Size"].str.replace("Varies with device","0")
df
```

```python
# appling cumalative summation
df["Total Rating"]=df["Rating"].cumsum().astype(float)
df
```

```python
# line plot
df.plot(x="Last Updated",y="Total Rating")
plt.ylabel("Total Rating")
```

```python
df.plot(x="Size",y="Rating",kind="scatter")
```

```python
df["Rating"].plot.hist()
```

```python
df.plot(x="Category",y="Rating",kind="bar")
```

```python
df["Total Rating"].plot(kind="bar",color=["salmon"])
```
